```python
from faker import Faker
import random
from generatorUtils import LabelTransformations
from itertools import permutations
```

```python
labels = ['day', 'day of the week', 'day name']
```

```python
class DayOfTheWeekGenerator:
    def __init__(self):
        
        self.labelTransformations = LabelTransformations()
        self.daysOfTheWeek = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

    def generateDaysOfTheWeek(self, labels):
        labels = list(set(self.labelTransformations(labels)))
        
        data_pairs =  list(permutations(self.daysOfTheWeek, 2))
        inputs = []
        outputs = []
        for inp, out in data_pairs:
            label = random.choice(labels)
            inp1 = self.labelTransformations.__all_words_capital__([inp])[0]
            out1 = self.labelTransformations.__all_words_capital__([out])[0]

            inp2 = self.labelTransformations.__all_words_start_capital__([inp])[0]
            out2 = self.labelTransformations.__all_words_start_capital__([out])[0]

            inp3 = self.labelTransformations.__all_words_small__([inp])[0]
            out3 = self.labelTransformations.__all_words_small__([out])[0]

            inp, out = random.choice([(inp1, out1), (inp2, out2), (inp3, out3)])
            inputs.append(label + ': ' + inp)
            outputs.append(out)
        return inputs, outputs
            
```

```python
DayOfTheWeekGenerator = DayOfTheWeekGenerator()
```

```python
input, output = DayOfTheWeekGenerator.generateDaysOfTheWeek(labels)
```

```python
import pandas as pd
res = pd.DataFrame(columns=['input_entity', 'output_entity'])
```

```python
res['input_entity'] = input
res['output_entity'] = output
```
